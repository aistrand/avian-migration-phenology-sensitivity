```python
list_rds_files = [i for i in os.walk(directory)][0][2]

print(len(list_rds_files))
```

```python
list_rds_files = [i for i in list_rds_files if '.rds' in i]

print(len(list_rds_files))
```

```python
list_rds_files_2002 = [i for i in list_rds_files if 'arrival_GAM_Hirundo_rustica_2002_' in i]

print(len(list_rds_files_2002))
```

```python
list_cells_2002 = [i.rpartition('_') for i in list_rds_files_2002]

print(len(list_cells_2002))

print(list_cells_2002[0:2])
```

```python
list_cells_2002 = [i[2] for i in list_cells_2002]

print(len(list_cells_2002))

print(list_cells_2002[0:2])
```

```python
list_cells_2002 = [i.rpartition('.') for i in list_cells_2002]

print(len(list_cells_2002))

print(list_cells_2002[0:2])
```

```python
list_cells_2002 = [i[0] for i in list_cells_2002]

print(len(list_cells_2002))

print(list_cells_2002[0:2])
```

```python
list_cells_2002 = [int(i) for i in list_cells_2002]

print(len(list_cells_2002))

print(list_cells_2002[0:2])
```

```python
df = pd.DataFrame(data={'cell': list_cells_2002})

print(len(df))

df.head(2)
```

```python
df['species'] = 'Hirundo_rustica'

df['year'] = 2002

df = df[['species', 'year', 'cell']]

print(len(df))

df.head(2)
```
