```python
list(df_gam.columns)
```

```python
list_columns_hm_iter = list(df_gam.columns[df_gam.columns.str.startswith('hm_iter_')])
```

```python
list_columns_hm_iter
```

```python
df_gam['mean_hm'] = df_gam[list_columns_hm_iter].mean(axis=1)
```

```python
df_gam.head(2)
```

```python
df_gam = df_gam.dropna(subset=['mean_hm'])

print(len(df_gam))
```

```python
df_gam.head(2)
```

Filters!

```python
df_gam = df_gam.rename(columns={'cell': 'grid_cell'})
```

```python
df_gam.head(2)
```

```python
df_merged = df.merge(df_gam, on=['year', 'grid_cell'])

print(len(df_merged))
```
