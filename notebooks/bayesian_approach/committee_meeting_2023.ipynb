```python
import matplotlib.pyplot as plt
```

```python
plt.plot([0, 1], [-1, 1])
plt.xticks([])
plt.yticks([-1, 0, 1], fontsize=15)
plt.xlabel('Body mass (g)', fontsize=20)
plt.ylabel('Sensitivity (days/°C)', fontsize=20)
plt.text(-0.3, 1, 'Lower', horizontalalignment='right', fontsize=15)
plt.text(-0.3, -1, 'Higher', horizontalalignment='right', fontsize=15)
plt.text(0, -1.3, 'Lower', fontsize=15)
plt.text(1, -1.3, 'Higher', horizontalalignment='right', fontsize=15)
plt.show()
```
