```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
constant = 10/7

df_routes = pd.DataFrame(np.array([
    [i*constant for i in [0.15, 0.2, 0.2, 0.3, 0.25]], 
    [i*constant for i in [0.15, 0.2, 0.3, 0.3, 0.2]], 
    [i*constant for i in [0.2, 0.3, 0.3, 0.25, 0.25]], 
    [i*constant for i in [0.25, 0.3, 0.3, 0.3, 0.25]], 
    [i*constant for i in [0.15, 0.2, 0.3, 0.3, 0.3]]]))
```

```python
ax = sns.heatmap(df_routes, cmap='Reds', vmin=0, vmax=1, xticklabels=False, yticklabels=False)
ax.set_xlabel('Longitude', fontsize=15)
ax.set_ylabel('Latitude', fontsize=15)
ax.set_title(
    'Proportion of occurrence of a given species on a given day in the spring in a given year', fontsize=15)
plt.text(-4, 1, 'A) Birds', fontsize=15)
plt.show()
```

```python
constant = 100

df_llj = pd.DataFrame(np.array([
    [i*constant for i in [0.15, 0.23, 0.3, 0.3, 0.25]], 
    [i*constant for i in [0.15, 0.25, 0.3, 0.3, 0.15]], 
    [i*constant for i in [0.25, 0.3, 0.3, 0.25, 0.25]], 
    [i*constant for i in [0.25, 0.3, 0.3, 0.3, 0.25]], 
    [i*constant for i in [0.15, 0.2, 0.3, 0.3, 0.3]]]))
```

```python
ax = sns.heatmap(df_llj, cmap='Reds', vmin=0, vmax=70, xticklabels=False, yticklabels=False)
ax.set_xlabel('Longitude', fontsize=15)
ax.set_ylabel('Latitude', fontsize=15)
ax.set_title(
    'Wind speed of the Great Plains low-level jet on a given day in the spring in a given year', fontsize=15)
plt.text(-4, 1, 'B) Great Plains low-level jet', fontsize=15)
plt.show()
```
